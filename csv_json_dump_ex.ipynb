## Writing the cmpltdProjects{} to a JSON file
    ## Once we have our cmpltdProjects dictionary processed and refined we now open a writer to output a JSON file
    ## Using the indent=4 argument, we will write the JSON to a file in a neat and pretty manner

```python
import json
def writeJson():
    #open up a writer:
    with open('projects_completed1.JSON', 'w') as f:
        #Creates the dump file to write to the indent creates a nice viewable file and the 
        #sort keys arg sorts the dict keys 
        json.dump(cmpltdProjects, f, indent=4, sort_keys=True)
        
        f.write('\n')
        f.close()
    print("JSON Written Successfully")
writeJson()
```
